```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Aditya\OneDrive\Desktop\Data_Science\crime_in_boston\crime.csv",encoding='utf-8')
```

```python
data.head()
```

```python
incidents = data[['INCIDENT_NUMBER','OFFENSE_CODE_GROUP','STREET','DISTRICT','OCCURRED_ON_DATE']]
incidents.to_csv(r'C:\Users\Aditya\OneDrive\Desktop\Data_Science\crime_in_boston\incidents.csv')
```

```python
incidents.head()
```

```python
incidents_offense = incidents.groupby('OFFENSE_CODE_GROUP')
```

```python
incidents_offense.groups
```

```python
l = list()
for offense in incidents_offense.groups:
    l.append(len(offense))
```

```python
l
```

```python
len(incidents_offense.groups.keys())
```

```python
len(l)
```

```python
Offense_Categorised = pd.DataFrame({'Offense':(i for i in incidents_offense.groups.keys()),
                                   'Number of Incidents':l})
```

```python
Offense_Categorised.to_csv(r'C:\Users\Aditya\OneDrive\Desktop\Data_Science\crime_in_boston\Categorised_Offence.csv')
```

```python
plt.figure(figsize=(30,25))
plt.barh(Offense_Categorised['Offense'],l)
plt.ylabel('OFFENSE')
plt.xlabel('NUMBER OF INCIDENTS')
plt.title('Offense Vs Number Of Such Incidents')
plt.savefig('OffenseVsNoOfSuchIncident.png')
```

```python
data.head()
```

```python
CrimesInEachYear=data.YEAR.value_counts()
type(CrimesInEachYear)
```

```python
CrimesInEachYear = pd.DataFrame({'Year': CrimesInEachYear.index,'Number Of Crimes': CrimesInEachYear})
```

```python
CrimesInEachYear.to_csv(r'C:\Users\Aditya\OneDrive\Desktop\Data_Science\crime_in_boston\Crime_In_Each_Year.csv')
```

```python
CrimesInEachYear.set_index((i for i in range(4)),inplace=True)
```

```python
CrimesInEachYear
```

```python
fig = plt.figure(figsize=(10,7))
plt.plot(CrimesInEachYear['Year'],CrimesInEachYear['Number Of Crimes'],color='green',marker='o',linewidth=2)
plt.xlabel('Year')
plt.ylabel('Number Of Crimes')
plt.title('Number Of Crimes Per Year')
Average_Crimes_Per_Year = CrimesInEachYear['Number Of Crimes'].mean()
fig.text(0.5,0.94,"\nAverage Crime Rate: " + str(Average_Crimes_Per_Year),ha='center')
plt.savefig(r'C:\Users\Aditya\OneDrive\Desktop\Data_Science\crime_in_boston\CrimePerYear.png')
plt.show()
```
